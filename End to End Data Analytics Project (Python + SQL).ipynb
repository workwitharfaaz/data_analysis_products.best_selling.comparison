```python
import pandas as pd
```

```python
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df.head(10)
```

```python
df['Ship Mode'].unique()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
```

```python
#now we will derive following columns - discount, sale price and profit
df["discount"]=df["list_price"]*df["discount_percent"]/100
df["sale_price"]=df["list_price"]-df["discount"]
df["profit"]=df["sale_price"]-df["cost_price"]
```

```python
df.head(1)
```

```python
pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```

```python
#Now we will load the data into SQL server using append
from sqlalchemy import create_engine

engine = create_engine("mysql+mysqlconnector://Arfaaz:********@localhost:3306/end2enddap")
conn=engine.connect()
df.to_sql('df_orders', con=conn, index=False, if_exists = 'append')
```
